```python
import numpy as np
from numba import njit, prange
import holoviews as hv

hv.extension('bokeh')
```

```python
@njit(cache=True)
def calculate_julia_point_njit(x, y, maxit, r=4):
    """
          ,        r=4.
        r,  ,   n->inf ,  z_n->inf.
         r  maxit ,  ,        r

    x, y :  ,      
    maxit :   
    r :   ,    

    return :  ,     ,  |z| > r
    """

    z = x + 1.0j * y  #  

    for lamb in range(maxit):
        z = z ** 4 + z ** 3 + 0.5 + 1.0j * -0.130649  #  
        if abs(z) > r:  #    
            return lamb  #        
    return maxit
```

```python
@njit(parallel=True, cache=True)
def create_julia_set_njit(xmin, xmax, ymin, ymax, nx, ny, maxit=10):
    """
        

    xmin, xmax, ymin, ymax : ,  ,      
    nx, ny :  ,      
    maxit :   

    return : numpy.array,       
    """

    dx = (xmax - xmin) / nx  #  
    dy = (ymax - ymin) / ny
    result = np.empty((ny + 1, nx + 1), dtype=np.int32)

    for iy in prange(ny + 1):  # 
        for ix in range(nx + 1):
            x = xmin + dx * ix  #  
            y = ymin + dy * iy
            result[iy, ix] = calculate_julia_point_njit(x, y, maxit) #  - 

    return result
```

```python
def create_julia_fractal_image(x_range, y_range):
    """
          holoviews

    x_range, y_range :  ,  

    return : holoviews.Image 
    """
    x0, x1 = x_range
    y0, y1 = y_range

    arr = create_julia_set_njit(x0, x1, -y1, -y0, w, h, maxit)
    return hv.Image(arr, bounds=(x0, y0, x1, y1))
```

```python
# 

w, h = 800, 800  #  
cx, cy = 0.1, 5.85  #  c   
maxit = 100  #   
xmin, xmax = -1.0, 1.0  #  ,    
ymin, ymax = -1.0, 1.0  #  ,    
```

```python
range_xy = hv.streams.RangeXY(x_range=(xmin, xmax), y_range=(ymin, ymax))

dmap = hv.DynamicMap(create_julia_fractal_image, streams=[range_xy])

dmap.opts(hv.opts.Image(height=w, width=h, cmap='reds', logz=False))
```
